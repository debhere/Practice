# EDA & Pre-Processing

```python
#Deleting un-important columns

ld_all = data_all.drop(columns=['ID', 'Amount.Funded.By.Investors'], axis=1)
ld_all.head()
```

```python
#Removing the % sign from Interest.Rate & Debt.To.Income.Ratio

#train_int = train.copy()
ld_all['Interest.Rate'] = ld_all['Interest.Rate'].str.replace('%', '')
ld_all['Debt.To.Income.Ratio'] = ld_all['Debt.To.Income.Ratio'].str.replace('%', '')
ld_all.head()
```

```python
#Converting the justified columns into numeric

#train_all = train_int.copy()

num_cols = ['Amount.Requested', 'Interest.Rate', 'Debt.To.Income.Ratio', 'Monthly.Income',
            'Open.CREDIT.Lines', 'Revolving.CREDIT.Balance', 'Inquiries.in.the.Last.6.Months']

for col in num_cols:
    ld_all[col] = pd.to_numeric(ld_all[col], errors='coerce')

ld_all.head(10)
```

```python
ld_all.info()
```

```python
ld_all.describe()
```

```python
for col in num_cols:
    m = ld_all[col].mean()
    ld_all[col].fillna(m, inplace=True)
```

```python
#Loan.Length

val = ld_all['Loan.Length'].value_counts()
print(val)
mode_loan = ld_all['Loan.Length'].mode()[0]
ld_all.loc[ld_all['Loan.Length'] == '.', 'Loan.Length'] = mode_loan
ld_all['Loan.Length'].fillna(mode_loan, inplace=True)
val = ld_all['Loan.Length'].value_counts()
print(val)
```

```python
tenure = pd.get_dummies(ld_all['Loan.Length'], prefix='Tenure', drop_first=True)
ld_all = pd.concat([ld_all, tenure], axis=1)
ld_all.drop('Loan.Length', axis=1, inplace=True)
ld_all.head()
```

```python
#Home.Ownership

val = ld_all['Home.Ownership'].value_counts()
print(val)
mode_own = ld_all['Home.Ownership'].mode()[0]
ld_all.loc[ld_all['Home.Ownership'] == 'OTHER', 'Home.Ownership'] = 'OWN'
ld_all.loc[ld_all['Home.Ownership'] == 'NONE', 'Home.Ownership'] = 'OWN'
ld_all['Home.Ownership'].fillna(mode_own, inplace=True)
val = ld_all['Home.Ownership'].value_counts()
print(val)
```

```python
home_own = pd.get_dummies(ld_all['Home.Ownership'], prefix='HO', drop_first=True)
ld_all = pd.concat([ld_all, home_own], axis=1)
ld_all.drop('Home.Ownership', axis=1, inplace=True)
ld_all.head()
```

```python
#FICO.Range

fico = ld_all['FICO.Range'].str.split('-', expand=True).astype(float)
ld_all['mean_fico'] = 0.5*(fico[0] + fico[1])
ld_all.drop('FICO.Range', axis=1, inplace=True)
ld_all.head()
```

```python
#Loan.Purpose

k = ld_all['Loan.Purpose'].value_counts()
k
```

```python
ld_all.loc[ld_all['Loan.Purpose'] == 'house', 'Loan.Purpose'] = 'home_improvement'
ld_all.loc[ld_all['Loan.Purpose'] == 'renewable_energy', 'Loan.Purpose'] = 'other'
ld_all.loc[ld_all['Loan.Purpose'] == 'educational', 'Loan.Purpose'] = 'other'
ld_all.loc[ld_all['Loan.Purpose'] == 'moving', 'Loan.Purpose'] = 'major_purchase'
ld_all.loc[ld_all['Loan.Purpose'] == 'vacation', 'Loan.Purpose'] = 'major_purchase'
ld_all.loc[ld_all['Loan.Purpose'] == 'car', 'Loan.Purpose'] = 'major_purchase'

purpose = pd.get_dummies(ld_all['Loan.Purpose'], prefix='purpose', drop_first=True)
ld_all = pd.concat([ld_all, purpose], axis=1)
ld_all.drop('Loan.Purpose', axis=1, inplace=True)

ld_all.shape
```

```python
ld_all.head()
```

```python
c = ld_all['Employment.Length'].value_counts()
c
```

```python
#Employment.Length


ld_all.loc[ld_all['Employment.Length'] == '10+ years', 'Employment.Length'] = '10 years'
ld_all.loc[ld_all['Employment.Length'] == '< 1 year', 'Employment.Length'] = '1 year'
ld_all.loc[ld_all['Employment.Length'] == '.', 'Employment.Length'] = '10 years'
```

```python
exp = pd.get_dummies(ld_all['Employment.Length'], prefix='Exp', drop_first=True)

ld_all = pd.concat([ld_all, exp], axis=1)

ld_all.drop('Employment.Length', axis=1, inplace=True)
ld_all.head()
```

```python
ld_all.drop('State', axis=1, inplace=True)

ld_all.info()
```

```python
ld_train = ld_all[ld_all['type'] == 'Train']
ld_test = ld_all[ld_all['type'] == 'Test']

X_train = ld_train.drop(columns = ['Interest.Rate', 'type'], axis=1)
y_train = ld_train['Interest.Rate']
X_test = ld_train.drop(columns = ['Interest.Rate', 'type'], axis=1)
```

```python
def metrics(scores):
    print('Scores: ', scores)
    print('Mean: ', scores.mean())
    print('Standard Deviation: ', scores.std())
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
```

```python
lin = LinearRegression()
lin.fit(X_train, y_train)
```

```python
lin_scores = cross_val_score(lin, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
lin_rmse = np.sqrt(-lin_scores)
metrics(lin_rmse)
```

```python
lasso = Lasso()
lasso.fit(X_train, y_train)
```

```python
lasso_scores = cross_val_score(lasso, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
lasso_rmse = np.sqrt(-lasso_scores)
metrics(lasso_rmse)
```

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
```

```python
ridge_scores = cross_val_score(ridge, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
ridge_rmse = np.sqrt(-ridge_scores)
metrics(ridge_rmse)
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
```

```python
tree_scores = cross_val_score(tree, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
tree_rmse = np.sqrt(-tree_scores)
metrics(tree_rmse)
```

```python
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
forest_scores = cross_val_score(forest, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
forest_rmse = np.sqrt(-forest_scores)
metrics(forest_rmse)
```

```python
param_list = {'n_estimators': [70, 80, 90],
              'max_depth': [7, 10, 12],
              'max_features': [2, 3, 4, 5]
             }

random_search = RandomizedSearchCV(forest, param_list, scoring="neg_mean_squared_error", cv=3)
random_search.fit(X_train, y_train)
```

```python
final_model = random_search.best_estimator_
```

```python
test_pred = final_model.predict(X_test)
```
